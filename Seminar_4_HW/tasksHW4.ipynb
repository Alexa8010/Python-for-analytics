Построим матрицу корреляций для определения наличия и силы корреляционных взаимосвязей имеющихся признаков:

```python
df_corr = dfH.corr(numeric_only=True)
df_corr = np.round(df_corr, 1)
df_corr
```

```python
plt.figure(figsize=(20, 8))

sns.set(font_scale=1.4)

sns.heatmap(df_corr, annot=True, linewidths=.5, cmap='coolwarm')

plt.title('Correlation matrix');
```

Получаем серию взаимосвязей для дальнейшего анализа, например:
Прямые сильные корреляции между:
- жилой площадью и стоимостью
- оценкой по 13-бальной шкале и стоимостью
- количеством ванных комнат и жилой прощадью
- квадратурой и ценой
Прямые умеренные между
- количесвом спален с жилой площадью
- количества этажей с количеством ванных комнат
Интересно, что широта обратно связана с индексом - особенности распределения нумерации.
